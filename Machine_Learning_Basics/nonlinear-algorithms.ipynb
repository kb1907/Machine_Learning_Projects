#### Decision trees are widely used models for classification and regression tasks. Essentially, they learn a hierarchy of if/else questions, leading to a decision.
#### Decision trees are flexible and powerful algorithms for both regression and classification problems.
#### To build a tree, the algorithm searches over all possible tests and finds the one that is most informative about the target variable.
#### Building a tree continues until all leaves are pure leads to models that are very complex and highly overfit to the training data. 
#### Overfitting can be prevented by stopping the building tree at the early stage or building the tree but removing the nodes which do not contain much inofrmation

#### Decision trees do not require to scaling, standartization or normalization of the features


<font color="blue" size=+0.5><b>Disadvantages</b></font>


#### Decision trees are prone to overfitting and high variance
#### Because of that, decision trees provide poor generalization

<font color="lightseagreen" size=+2.5><b>Random Forests</b></font>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" style="color:white" data-toggle="popover">Table of Contents</a>


#### As we have mentioned, one of the main disadvantages of decision trees is that they have high variance, tend to overfit and as an expected result we will have poor accuracy and poor generalization performance. 

#### To reduce high varaiance of the decision trees,ensemble methods (Bootstrap Aggregation, etc.) are used.

#### **Ensemble**: Building a strong model by using a ensemble of weaker models.  

#### In the Random forest algorithm,  to reduce the variance, the sub-trees are learned so that the resulting predictions from all of the subtrees have less correlation.

#### Basically, Random forest algorithm build many trees,even though each of them overfit the data, takes average of the their results to reduce overfitting.

Image Credit : https://www.pcigeomatics.com/

<font color="blue" size=+0.5><b>Advantages</b></font>

#### Random Forests reduces the overfitting and high variance.
#### The model works with both regression and the classification problems.
#### The Random Forests model does not require normalizing of data.


<font color="blue" size=+0.5><b>Disadvantages</b></font>
#### The model requires more time to train and needs more computational power.
#### The model is not easy to interpret
#### The model does not provide high level performance on the high dimensional data (such as text data, etc.)

<font color="lightseagreen" size=+2.5><b>Naive Bayes</b></font>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" style="color:white" data-toggle="popover">Table of Contents</a>


#### **Naive** : It is called Naive because it assumes that occurance of the each input is independent of the occurrence of other inputs. 

#### **Bayes** : It comes from Bayes's Theorem (<u>**see below formula**</u>)

#### Naive Bayes algorithm is a classification algoritmh for both binary and multiclass classification problems.

#### Naive Bayes algorithm is generally used in text analysis.

#### Naive Bayes learns parameters by looking each input seperately and collects the simple statistics from each feature. Which basically means, model uses prior knowledge about the problem.

#### Naive Bayes is very fast to learn from the training data.

#### Naive Bayes is used for **Text Data Classification** (such as sentiment analysis, etc.), **Credit Scoring**,**Spam Filtering**, **Medical Data Classification** and **Real-Time Prediction**.

#### Three kinds of naive Bayes classifiers:

#### GaussianNB, BernoulliNB, and MultinomialNB

#### **GaussianNB** can be used with any data.

#### **BernoulliNB**, and **MultinomialNB** are mostly used with text data classification.


<font color="blue" size=+0.5><b>Bayes's Theorem </b></font>

#### **P(A|B)** : Posterior probability. 
#### **P(A)** :  Prior probability
#### **P(B|A)** : Probability of data d given that the hypothesis A was true.
#### **P(B)** :  Probability of the data


#### We try to get posterior probability by using prior probability.

<font color="blue" size=+0.5><b>Advantages</b></font>

#### Naive Bayes models are fast to train and to predict 
#### Naive Bayes models' training procedures are easy to understand. 
#### The model works very well with high-dimensional sparse data and it is relatively robust to the parameters. 
#### Naive Bayes models are great baseline models and are often used on very large datasets, where training even a linear model might take too long.


<font color="blue" size=+0.5><b>Disadvantages</b></font>
#### Naive Bayes models assumes that all input variables are unrelated for that reason model can't learn from the relations between the input variables.
#### <u> **Zero Frequency** </u>: Let's imagine, we have a categorical variable with three categories. (category A, category B, category C). In the training data we have only two categories of this categorical variable (category A, category B) and one of the category (category C) of the categorical variable wasn't present in the training data. The Naive Bayes model will assign it zero probability. But in the test data we have that category of the categorical variable. Since Naive Bayes model assign zero probability to category C, model won‚Äôt be able to make any predictions in this regard.

<font color="lightseagreen" size=+2.5><b>KNN (k- Nearest Neighbors)</b></font>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" style="color:white" data-toggle="popover">Table of Contents</a>


#### The KNN algorithm is one of the most simple and still very effective algorithm.

#### When we have new data point, our model makes predictions by searching through the entire training set for the similar instances (K neighbors).

#### 


![](https://miro.medium.com/max/591/1*kCqervQNQ5fGDfkFwrMzRQ.png)

Image Credit: miro.medium.com

<font color="blue" size=+0.5><b>Advantages</b></font>
#### KNN model is simple. Only requires to adjust the number of neighbors, k and  which the distance metric to be used, then ready to go.
#### KNN models are fast to train. 
#### KNN models are easy to understand. 
#### KNN models don't make any assumption about the data. For example, linear regression expects linear relation between input and target variables, but KNN models do not require any assumption about the data and work with it. 


<font color="blue" size=+0.5><b>Disadvantages</b></font>
#### KNN models needs preprocessing the data (scaling and balancing).
#### KNN models are sensitive to the outliers
#### KNN models can't deal with the missing values.
#### Requires large memory to store the training data.
#### Since it requires a lot of memory and its' inability to deal with many features, KNN models are not often used in the practice.

<font color="lightseagreen" size=+2.5><b>Support Vector Machines (SVM)</b></font>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" style="color:white" data-toggle="popover">Table of Contents</a>




#### Even though SVW is used for both regression and classification problems, it is primarily used for classification problems. 

#### The main goal of the SVM algorithm is to create the best line or decision boundary. This best decision boundary is called a **hyperplane**.

#### Support vectors in this sense, are the data points that are closer to a hyperplane.

#### SVM aims to maximize the distance (margin) between each of the support vectors. 

#### SVM works well with the high dimensional data (Text data, gene data, image data, etc.)

![](https://static.javatpoint.com/tutorial/machine-learning/images/support-vector-machine-algorithm.png)

Image Credit : www.javatpoint.com

<font color="blue" size=+0.5><b>Advantages</b></font>

#### SVM draws boundary between the data points are not linearly separable.
#### SVM  works effectively in the higher dimension.
#### SVM is not affected by the outliers, model's only concern is support vectors.
 


<font color="blue" size=+0.5><b>Disadvantages</b></font>
#### SVM is slow when dealing with the larger dataset.
#### Even though SVM works well with the seperable classes, if classes are overlapped, SVM shows poor performance.
